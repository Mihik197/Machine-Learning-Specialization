```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_array = np.array([1, 2, 3])
exp_array = np.exp(input_array)  # e^x

print("Input to exp:", input_array)
print("Output of exp:", exp_array)

# single input
input_val = 1  
exp_val = np.exp(input_val)

print("\nInput to exp:", input_val)
print("Output of exp:", exp_val)
```

```python
# sigmoid function
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
# trying this function
z_temp = np.arange(-10, 11)  # -10 to 10 (excluding 11)

y_temp = sigmoid(z_temp)

# neatly printing both arrays side by side
np.set_printoptions(precision=3)
print("Input(z), Output (sigmoid(z))")
print(np.c_[z_temp, y_temp])

# output values ranging from 0 to 1 as expected
```

```python
# z_temp vs sigmoid output
fig, ax = plt.subplots(1, 1, figsize=(5, 3))  # 1, 1 stands for the grid layout of the subplots. number of rows and columns
ax.plot(z_temp, y_temp, c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel("sigmoid(z)")
ax.set_xlabel("z")
plt.show()
```

```python
# decision boundaries
```

```python
# helper function
def plot_data(X, y, ax, pos_label="y=1", neg_label="y=0", s=80, loc='best' ):
    """ plots logistic data with two axis """
    # Find Indices of Positive and Negative Examples
    pos = y == 1
    neg = y == 0
    pos = pos.reshape(-1,)  #work with 1D or 1D y vectors
    neg = neg.reshape(-1,)

    # Plot examples
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', s=s, c = 'red', label=pos_label)
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=s, label=neg_label, facecolors='none', edgecolors='blue', lw=3)
    ax.legend(loc=loc)
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  # 2 features per input
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 

fig, ax = plt.subplots(1, 1, figsize=(4, 4))
plot_data(X, y, ax)

ax.axis([0, 4, 0, 3.5])  # minimum and maximum limits for x and y axes
ax.set_xlabel('x0')
ax.set_ylabel('x1')
plt.show()
```

example case:  

Our logistic regression model has the form  
$f(\mathbf{x}) = g(-3 + x_0+x_1)$

this model predicts $y=1$ if $-3 + x_0+x_1 >= 0$

we start by plotting $-3 + x_0+x_1 = 0$, which is equivalent to $x_1 = 3 - x_0$

```python
x0 = np.arange(0,6)

x1 = 3 - x0

fig,ax = plt.subplots(1,1,figsize=(5,4))
ax.plot(x0, x1, c='b')  # plotting the decision boundary
ax.axis([0, 4, 0, 3.5])

# filling the region below the line with color
ax.fill_between(x0, x1, alpha=0.2)

# original data for reference
plot_data(X, y, ax)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
plt.show()
```

```python
# any point above the line is classified as y=1 and below is y=0
```

### Logistic Loss function

```python
# since logistic regression has a non-linear component, the sigmoid function
# we can't use the square error cost function
# as it creates a non-convex, so we can get stuck in a local minima
```



The loss function above can be rewritten to be easier to implement.
    $$loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = (-y^{(i)} \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)$$

```python
# same data as before
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  # 2 features per input
y_train = np.array([0, 0, 0, 1, 1, 1]) 
```

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X_train, y_train, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.show()
```

```python
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]

    cost = 0

    for i in range(m):
        z_i = np.dot(X[i], w) + b
        f_wb_i = sigmoid(z_i)
        cost += (-y[i] * np.log(f_wb_i)) - ((1 - y[i])*np.log(1 - f_wb_i))

    cost = cost / m
    return cost
```

```python
w_temp = np.array([1, 1])
b_temp = -3
print(compute_cost_logistic(X_train, y_train, w_temp, b_temp))
```

```python
# purposely choosing a worse bias to see if our loss is working as intended
w_array1 = np.array([1,1])
b_1 = -3
w_array2 = np.array([1,1])
b_2 = -4

print("Cost for b = -3 : ", compute_cost_logistic(X_train, y_train, w_array1, b_1))
print("Cost for b = -4 : ", compute_cost_logistic(X_train, y_train, w_array2, b_2))
```
