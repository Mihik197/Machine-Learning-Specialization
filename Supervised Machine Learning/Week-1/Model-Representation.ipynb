```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])  # size in thousand sq feet
y_train = np.array([300.0, 500.0])  # cost in thousand dollars

print(x_train)
print(y_train)
```

```python
w = 200
b = 100
print(f'w: {w}')
print(f'b: {b}')
```

```python
# Computes the prediction of a linear model
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, tmp_f_wb, c='b', label='Our prediction')  # model prediction
# label -> creates a legend key table in the corner

plt.scatter(x_train, y_train, marker='x', c='r', label="Actual Data")  # actual data points

plt.title('Housing Prices')
plt.ylabel('Price (in 1000 dollars)')
plt.xlabel('Size (in 1000 sq feet)')
plt.legend()
plt.show()
```

```python
# prediction

x_i = 1.2  # for (1200 sq feet house)
cost_1200 = w * x_i + b
print(f'Cost of 1200 sq feet house: ${cost_1200 * 1000:,.0f}')
```

```python
# cost function
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i]) ** 2
    total_cost = (cost / (2 * m))  # we use 1/2m instead of 1/m to average, as it helps us when taking the derivative of the loss

    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    
    wrt_w = 0
    wrt_b = 0

    for i in range(m):
        f_wb = w * x[i] + b
        w_i = (f_wb - y[i]) * x[i]
        b_i = (f_wb - y[i])
        wrt_w += w_i 
        wrt_b += b_i

    wrt_w = wrt_w / m
    wrt_b = wrt_b / m

    return wrt_w, wrt_b
```

```python
def gradient_descent(x, y, w_in, b_in, lr, num_iters, cost_function, gradient_function):
    J = []  # to store history of costs
    p = []  # to store history of w and b
    
    w = w_in
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(x, y, w, b)

        w = w - lr * dj_dw
        b = b - lr * dj_db

        J.append(cost_function(x, y, w, b))
        p.append([w, b])

        if i % 1000 == 0:
            print(f'Cost on iteration {i + 1}: {J[-1]:.2f}')
            print(f'w: {w:.2f}, b: {b:.2f}')

    return w, b, J, p
```

```python
lr = 1.0e-2
w_init = 0
b_init = 0
iterations = 10000

w_final, b_final, J_hist, p_hist = gradient_descent(x_train, y_train, w_init, b_init, lr, iterations, compute_cost, compute_gradient)
print(f'after gradient descent:')
print(f'w: {w_final:.2f}, b: {b_final:.2f}')
```
